```python
# depression_education_analysis.py

import pandas as pd
import matplotlib.pyplot as plt

# Sample data 
data = {
    'Year': [2012, 2013, 2014, 2015, 2016, 2017, 2018],
    'Strata': ['Education'] * 7,
    'Strata Name': ['High School Graduate or GED Certificate'] * 7,
    'Frequency': [337, 361, 209, 328, 263, 265, 337],
    'Weighted Frequency': [665672, 841995, 625805, 670489, 833318, 1297640, 830587],
    'Percent': [11.44, 14.28, 12.80, 12.47, 14.33, 22.45, 15.53],
    'Lower 95% CL': [9.98, 12.50, 10.14, 10.28, 11.26, 16.90, 11.40],
    'Upper 95% CL': [12.89, 16.06, 15.47, 14.66, 17.41, 28.01, 19.65]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate error margins for error bars
df['Error_Low'] = df['Percent'] - df['Lower 95% CL']
df['Error_High'] = df['Upper 95% CL'] - df['Percent']

# Plotting
plt.figure(figsize=(10, 6))
plt.errorbar(
    df['Year'], df['Percent'],
    yerr=[df['Error_Low'], df['Error_High']],
    fmt='-o', capsize=5, color='teal', ecolor='lightgray', elinewidth=3
)

plt.title('Depression Prevalence Among High School Graduates (2012-2018)')
plt.xlabel('Year')
plt.ylabel('Percent with Depressive Disorders (%)')
plt.grid(True)
plt.xticks(df['Year'])
plt.tight_layout()
plt.savefig('depression_trend.png')
plt.show()

# Summary Statistics
summary = df[['Year', 'Percent', 'Lower 95% CL', 'Upper 95% CL']]
print("\nSummary of Depression Percentages and Confidence Intervals:")
print(summary.to_string(index=False))
```
